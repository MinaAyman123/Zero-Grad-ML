```python
# make sure both columns are datetime type
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# calculate delivery time
df['Delivery_Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
def hist_numeric(df,col) : 
    print(df[col].value_counts())
    sns.histplot(df[col])
    plt.show()

for item in ['Quantity','Discount','Delivery_Days'] :
    hist_numeric(df,item)
```

```python
def count_cat(df,col) : 
    print(df[col].value_counts())
    sns.countplot(x=df[col])
    plt.show()

for item in ['Segment','Category']:
    count_cat(df,item)
```

```python
def bar(df,col):
    print(df.groupby(col)['Profit'].mean())
    sns.barplot(x=col , y='Profit' , data=df)
    plt.axhline(df['Profit'].mean() , color='red')
    plt.show()

for i in [ 'Category','Segment','Ship Mode']:
    bar(df,i)
```
